```python
import os
import glob
from collections import defaultdict
import datetime
import csv # CSV ì²˜ë¦¬ë¥¼ ìœ„í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì¶”ê°€

# --- ì‚¬ìš©ì ì„¤ì • (í•„ìˆ˜ ìˆ˜ì •) ---
class USER_CONFIG:
    # 1. 'car' í´ë˜ìŠ¤ì˜ ID (0ë¶€í„° ì‹œì‘)
    CAR_CLASS_ID = 2

    # 2. YOLO ì˜ˆì¸¡ ê²°ê³¼ í´ë” ê²½ë¡œ (ê°€ì¥ ìµœê·¼ì— ìƒì„±ëœ í´ë” ì´ë¦„ìœ¼ë¡œ ìˆ˜ì •)
    # ì˜ˆ: "/content/runs/detect/predict2"
    PREDICT_RESULTS_DIR = "/content/drive/MyDrive/result/predict7"

    # 3. ëª¨ë¸ ì´ë¦„ (ì¶œë ¥ íŒŒì¼ëª…ì— ì‚¬ìš©)
    MODEL_NAME = "yolov5n"

    MODEL_CONF = "default"
# -----------------------------

def check_car_presence_and_save(config: USER_CONFIG):
    """
    YOLO ì˜ˆì¸¡ ë ˆì´ë¸” íŒŒì¼ì„ ë¶„ì„í•˜ì—¬ ê° ì´ë¯¸ì§€ì— 'car'ê°€ ìˆëŠ”ì§€ í™•ì¸í•˜ê³ 
    ê²°ê³¼ë¥¼ ì½˜ì†”, TXT íŒŒì¼, CSV íŒŒì¼ë¡œ ì €ì¥í•©ë‹ˆë‹¤.
    """
    results_dir = config.PREDICT_RESULTS_DIR
    target_class_id = config.CAR_CLASS_ID
    model_name = config.MODEL_NAME
    model_conf = config.MODEL_CONF

    labels_path = os.path.join(results_dir, "labels")
    if not os.path.isdir(labels_path):
        print(f"âš ï¸ ì˜¤ë¥˜: ë ˆì´ë¸” í´ë”ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {labels_path}")
        print("ê²½ë¡œì™€ YOLO predict ì‹¤í–‰ ì—¬ë¶€ë¥¼ í™•ì¸í•˜ì„¸ìš”.")
        return

    label_files = glob.glob(os.path.join(labels_path, "*.txt"))
    if not label_files:
        print(f"âš ï¸ ì˜¤ë¥˜: {labels_path} í´ë”ì— .txt ë ˆì´ë¸” íŒŒì¼ì´ ì—†ìŠµë‹ˆë‹¤.")
        return

    presence_summary = {}

    # ê²°ê³¼ë¥¼ ì €ì¥í•  ë¬¸ìì—´ ë²„í¼ ì´ˆê¸°í™” (ì½˜ì†”/TXTìš©)
    output_buffer = []

    # --- 1. ë¶„ì„ ë° ë°ì´í„° ìˆ˜ì§‘ ---
    output_buffer.append("-" * 50)
    output_buffer.append(f"ğŸ•µï¸â€â™€ï¸ ì´ {len(label_files)}ê°œ ì´ë¯¸ì§€ì˜ 'car' ì¡´ì¬ ì—¬ë¶€ ë¶„ì„ ì¤‘...")
    output_buffer.append(f"ğŸ” ëŒ€ìƒ í´ë˜ìŠ¤: 'car' (ID: {target_class_id})")
    output_buffer.append("-" * 50)

    # ëª¨ë“  ë ˆì´ë¸” íŒŒì¼ ìˆœíšŒ
    for file_path in label_files:
        # YOLOëŠ” .txt íŒŒì¼ì„ ì´ë¯¸ì§€ íŒŒì¼ëª… (ì˜ˆ: .jpg)ìœ¼ë¡œ ê°„ì£¼í•˜ê³  ì²˜ë¦¬í•©ë‹ˆë‹¤.
        # ë”°ë¼ì„œ .txt í™•ì¥ìë¥¼ .jpgë¡œ ëŒ€ì²´í•˜ì—¬ ì›ë³¸ ì´ë¯¸ì§€ íŒŒì¼ëª…ì„ ì¶”ì •í•©ë‹ˆë‹¤.
        base_name = os.path.basename(file_path).replace('.txt', '.jpg')
        has_car = False

        try:
            with open(file_path, 'r') as f:
                for line in f:
                    parts = line.strip().split()

                    if len(parts) >= 1:
                        try:
                            class_id = int(parts[0])
                            if class_id == target_class_id:
                                has_car = True
                                break # 'car'ê°€ í•˜ë‚˜ë¼ë„ ìˆìœ¼ë©´ ì¦‰ì‹œ ì¤‘ë‹¨
                        except ValueError:
                            # ì²« ë²ˆì§¸ ë¶€ë¶„ì´ ìˆ«ìê°€ ì•„ë‹Œ ê²½ìš° ê±´ë„ˆë›°ê¸°
                            continue

        except Exception as e:
            output_buffer.append(f"íŒŒì¼ {file_path} ì²˜ë¦¬ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")
            continue

        presence_summary[base_name] = 'O' if has_car else 'X'

    # --- 2. í†µê³„ ë° ë¹„ìœ¨ ê³„ì‚° ---
    total_images = len(label_files)
    present_images = sum(1 for status in presence_summary.values() if status == 'O')
    absent_images = total_images - present_images

    if total_images > 0:
        presence_percentage = (present_images / total_images) * 100
    else:
        presence_percentage = 0.0

    # ë‚ ì§œ ë° ì‹œê°„ í¬ë§· ìƒì„±
    timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")

    # íŒŒì¼ëª… ì •ì˜
    summary_txt_filename = f"{model_name}_{model_conf}_{timestamp}_performance_summary.txt" # ì„±ëŠ¥ ìš”ì•½ TXT íŒŒì¼
    detail_csv_filename = f"{model_name}_{model_conf}_{timestamp}_car_presence_detail.csv" # ìƒì„¸ ë‚´ì—­ CSV íŒŒì¼

    # --- 3. CSV íŒŒì¼ë¡œ ìƒì„¸ ë‚´ì—­ ì €ì¥ ---
    # CSV íŒŒì¼ì—ëŠ” ì´ë¯¸ì§€ íŒŒì¼ëª…ê³¼ ì¡´ì¬ ì—¬ë¶€ ('O'/'X')ë¥¼ ì €ì¥í•©ë‹ˆë‹¤.
    try:
        with open(detail_csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
            fieldnames = ['image_filename', 'car_presence']
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

            writer.writeheader()
            for filename, status in presence_summary.items():
                writer.writerow({'image_filename': filename, 'car_presence': status})

        output_buffer.append(f"\nâœ… ìƒì„¸ ë¶„ì„ ê²°ê³¼ê°€ ì„±ê³µì ìœ¼ë¡œ '{detail_csv_filename}' íŒŒì¼ì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
        output_buffer.append(f"CSV íŒŒì¼ ê²½ë¡œ: {os.path.abspath(detail_csv_filename)}")

    except Exception as e:
        output_buffer.append(f"\nâŒ CSV íŒŒì¼ ì €ì¥ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")

    # --- 4. ì½˜ì†” ì¶œë ¥ ë° TXT íŒŒì¼ í¬ë§·íŒ… (ë²„í¼ì— ì¶”ê°€) ---
    # ìƒì„¸ ë‚´ì—­ í…Œì´ë¸” (CSVì™€ ë™ì¼í•œ ë‚´ìš©)
    output_buffer.append("\n" + "=" * 60)
    output_buffer.append(f"ğŸ‰ 'car' í´ë˜ìŠ¤ (ID: {target_class_id}) ì¡´ì¬ ì—¬ë¶€ ìƒì„¸ ìš”ì•½ (CSVì— ì €ì¥ë¨)")
    output_buffer.append("=" * 60)
    output_buffer.append(f"| ì´ë¯¸ì§€ íŒŒì¼ëª… | ì¡´ì¬ ì—¬ë¶€ |")
    output_buffer.append("| :--- | :---: |")

    # ì½˜ì†”/TXT ì¶œë ¥ ì‹œ ê°€ë…ì„±ì„ ìœ„í•´ ì¼ë¶€ë§Œ ì¶œë ¥í•˜ê±°ë‚˜, ì „ì²´ë¥¼ ì¶œë ¥í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
    # ì—¬ê¸°ì„œëŠ” ì œê³µí•´ì£¼ì‹  ì˜ˆì‹œì™€ ë™ì¼í•˜ê²Œ ì „ì²´ ëª©ë¡ì„ ì¶œë ¥í•©ë‹ˆë‹¤.
    for filename, status in presence_summary.items():
        output_buffer.append(f"| {filename} | {status} |")

    # ìµœì¢… ì„±ëŠ¥ ìš”ì•½
    output_buffer.append("=" * 60)
    output_buffer.append(f"ğŸ“Š ìµœì¢… ì„±ëŠ¥ ìš”ì•½ (ëª¨ë¸: {model_name})")
    output_buffer.append("-" * 60)
    output_buffer.append(f"ì´ ì´ë¯¸ì§€ ìˆ˜: {total_images}ê°œ")
    output_buffer.append(f"âœ… 'car' ìˆìŒ (O): {present_images}ê°œ")
    output_buffer.append(f"âŒ 'car' ì—†ìŒ (X): {absent_images}ê°œ")
    output_buffer.append(f"â¡ï¸ 'car'ê°€ íƒì§€ëœ ì´ë¯¸ì§€ ë¹„ìœ¨: **{presence_percentage:.2f}%**")
    output_buffer.append("-" * 60)

    # --- 5. ì½˜ì†” ì¶œë ¥ ë° ì„±ëŠ¥ ìš”ì•½ TXT íŒŒì¼ ì €ì¥ ---
    final_output = "\n".join(output_buffer)

    # ì½˜ì†”ì— ì¶œë ¥
    print(final_output)

    # TXT íŒŒì¼ë¡œ ì €ì¥ (ì„±ëŠ¥ ìš”ì•½ í¬í•¨)
    try:
        with open(summary_txt_filename, 'w', encoding='utf-8') as f:
            f.write(final_output)
        print(f"\nâœ… ë¶„ì„ ê²°ê³¼ ìš”ì•½ì´ ì„±ê³µì ìœ¼ë¡œ '{summary_txt_filename}' íŒŒì¼ì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
        print(f"TXT íŒŒì¼ ê²½ë¡œ: {os.path.abspath(summary_txt_filename)}")
    except Exception as e:
        print(f"âŒ TXT íŒŒì¼ ì €ì¥ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")


# í•¨ìˆ˜ ì‹¤í–‰
if __name__ == "__main__":
    check_car_presence_and_save(USER_CONFIG())
```

```python
import os
import glob
from collections import defaultdict
import datetime
import csv # CSV ì²˜ë¦¬ë¥¼ ìœ„í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì¶”ê°€

# --- ì‚¬ìš©ì ì„¤ì • (í•„ìˆ˜ ìˆ˜ì •) ---
class USER_CONFIG:
    # 1. 'car' í´ë˜ìŠ¤ì˜ ID (0ë¶€í„° ì‹œì‘)
    CAR_CLASS_ID = 2

    # 2. YOLO ì˜ˆì¸¡ ê²°ê³¼ í´ë” ê²½ë¡œ (ê°€ì¥ ìµœê·¼ì— ìƒì„±ëœ í´ë” ì´ë¦„ìœ¼ë¡œ ìˆ˜ì •)
    # ì˜ˆ: "/content/runs/detect/predict2"
    PREDICT_RESULTS_DIR = "/content/drive/MyDrive/result/predict12"

    # 3. ëª¨ë¸ ì´ë¦„ (ì¶œë ¥ íŒŒì¼ëª…ì— ì‚¬ìš©)
    MODEL_NAME = "yolo11n"

    MODEL_CONF = "0.05"
# -----------------------------

def check_car_presence_and_save(config: USER_CONFIG):
    """
    YOLO ì˜ˆì¸¡ ë ˆì´ë¸” íŒŒì¼ì„ ë¶„ì„í•˜ì—¬ ê° ì´ë¯¸ì§€ì— 'car'ê°€ ìˆëŠ”ì§€ í™•ì¸í•˜ê³ 
    ê²°ê³¼ë¥¼ ì½˜ì†”, TXT íŒŒì¼, CSV íŒŒì¼ë¡œ ì €ì¥í•©ë‹ˆë‹¤.
    """
    results_dir = config.PREDICT_RESULTS_DIR
    target_class_id = config.CAR_CLASS_ID
    model_name = config.MODEL_NAME
    model_conf = config.MODEL_CONF

    labels_path = os.path.join(results_dir, "labels")
    if not os.path.isdir(labels_path):
        print(f"âš ï¸ ì˜¤ë¥˜: ë ˆì´ë¸” í´ë”ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {labels_path}")
        print("ê²½ë¡œì™€ YOLO predict ì‹¤í–‰ ì—¬ë¶€ë¥¼ í™•ì¸í•˜ì„¸ìš”.")
        return

    label_files = glob.glob(os.path.join(labels_path, "*.txt"))
    if not label_files:
        print(f"âš ï¸ ì˜¤ë¥˜: {labels_path} í´ë”ì— .txt ë ˆì´ë¸” íŒŒì¼ì´ ì—†ìŠµë‹ˆë‹¤.")
        return

    presence_summary = {}

    # ê²°ê³¼ë¥¼ ì €ì¥í•  ë¬¸ìì—´ ë²„í¼ ì´ˆê¸°í™” (ì½˜ì†”/TXTìš©)
    output_buffer = []

    # --- 1. ë¶„ì„ ë° ë°ì´í„° ìˆ˜ì§‘ ---
    output_buffer.append("-" * 50)
    output_buffer.append(f"ğŸ•µï¸â€â™€ï¸ ì´ {len(label_files)}ê°œ ì´ë¯¸ì§€ì˜ 'car' ì¡´ì¬ ì—¬ë¶€ ë¶„ì„ ì¤‘...")
    output_buffer.append(f"ğŸ” ëŒ€ìƒ í´ë˜ìŠ¤: 'car' (ID: {target_class_id})")
    output_buffer.append("-" * 50)

    # ëª¨ë“  ë ˆì´ë¸” íŒŒì¼ ìˆœíšŒ
    for file_path in label_files:
        # YOLOëŠ” .txt íŒŒì¼ì„ ì´ë¯¸ì§€ íŒŒì¼ëª… (ì˜ˆ: .jpg)ìœ¼ë¡œ ê°„ì£¼í•˜ê³  ì²˜ë¦¬í•©ë‹ˆë‹¤.
        # ë”°ë¼ì„œ .txt í™•ì¥ìë¥¼ .jpgë¡œ ëŒ€ì²´í•˜ì—¬ ì›ë³¸ ì´ë¯¸ì§€ íŒŒì¼ëª…ì„ ì¶”ì •í•©ë‹ˆë‹¤.
        base_name = os.path.basename(file_path).replace('.txt', '.jpg')
        has_car = False

        try:
            with open(file_path, 'r') as f:
                for line in f:
                    parts = line.strip().split()

                    if len(parts) >= 1:
                        try:
                            class_id = int(parts[0])
                            if class_id == target_class_id:
                                has_car = True
                                break # 'car'ê°€ í•˜ë‚˜ë¼ë„ ìˆìœ¼ë©´ ì¦‰ì‹œ ì¤‘ë‹¨
                        except ValueError:
                            # ì²« ë²ˆì§¸ ë¶€ë¶„ì´ ìˆ«ìê°€ ì•„ë‹Œ ê²½ìš° ê±´ë„ˆë›°ê¸°
                            continue

        except Exception as e:
            output_buffer.append(f"íŒŒì¼ {file_path} ì²˜ë¦¬ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")
            continue

        presence_summary[base_name] = 'O' if has_car else 'X'

    # --- 2. í†µê³„ ë° ë¹„ìœ¨ ê³„ì‚° ---
    total_images = len(label_files)
    present_images = sum(1 for status in presence_summary.values() if status == 'O')
    absent_images = total_images - present_images

    if total_images > 0:
        presence_percentage = (present_images / total_images) * 100
    else:
        presence_percentage = 0.0

    # ë‚ ì§œ ë° ì‹œê°„ í¬ë§· ìƒì„±
    timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")

    # íŒŒì¼ëª… ì •ì˜
    summary_txt_filename = f"{model_name}_{model_conf}_{timestamp}_performance_summary.txt" # ì„±ëŠ¥ ìš”ì•½ TXT íŒŒì¼
    detail_csv_filename = f"{model_name}_{model_conf}_{timestamp}_car_presence_detail.csv" # ìƒì„¸ ë‚´ì—­ CSV íŒŒì¼

    # --- 3. CSV íŒŒì¼ë¡œ ìƒì„¸ ë‚´ì—­ ì €ì¥ ---
    # CSV íŒŒì¼ì—ëŠ” ì´ë¯¸ì§€ íŒŒì¼ëª…ê³¼ ì¡´ì¬ ì—¬ë¶€ ('O'/'X')ë¥¼ ì €ì¥í•©ë‹ˆë‹¤.
    try:
        with open(detail_csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
            fieldnames = ['image_filename', 'car_presence']
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

            writer.writeheader()
            for filename, status in presence_summary.items():
                writer.writerow({'image_filename': filename, 'car_presence': status})

        output_buffer.append(f"\nâœ… ìƒì„¸ ë¶„ì„ ê²°ê³¼ê°€ ì„±ê³µì ìœ¼ë¡œ '{detail_csv_filename}' íŒŒì¼ì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
        output_buffer.append(f"CSV íŒŒì¼ ê²½ë¡œ: {os.path.abspath(detail_csv_filename)}")

    except Exception as e:
        output_buffer.append(f"\nâŒ CSV íŒŒì¼ ì €ì¥ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")

    # --- 4. ì½˜ì†” ì¶œë ¥ ë° TXT íŒŒì¼ í¬ë§·íŒ… (ë²„í¼ì— ì¶”ê°€) ---
    # ìƒì„¸ ë‚´ì—­ í…Œì´ë¸” (CSVì™€ ë™ì¼í•œ ë‚´ìš©)
    output_buffer.append("\n" + "=" * 60)
    output_buffer.append(f"ğŸ‰ 'car' í´ë˜ìŠ¤ (ID: {target_class_id}) ì¡´ì¬ ì—¬ë¶€ ìƒì„¸ ìš”ì•½ (CSVì— ì €ì¥ë¨)")
    output_buffer.append("=" * 60)
    output_buffer.append(f"| ì´ë¯¸ì§€ íŒŒì¼ëª… | ì¡´ì¬ ì—¬ë¶€ |")
    output_buffer.append("| :--- | :---: |")

    # ì½˜ì†”/TXT ì¶œë ¥ ì‹œ ê°€ë…ì„±ì„ ìœ„í•´ ì¼ë¶€ë§Œ ì¶œë ¥í•˜ê±°ë‚˜, ì „ì²´ë¥¼ ì¶œë ¥í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
    # ì—¬ê¸°ì„œëŠ” ì œê³µí•´ì£¼ì‹  ì˜ˆì‹œì™€ ë™ì¼í•˜ê²Œ ì „ì²´ ëª©ë¡ì„ ì¶œë ¥í•©ë‹ˆë‹¤.
    for filename, status in presence_summary.items():
        output_buffer.append(f"| {filename} | {status} |")

    # ìµœì¢… ì„±ëŠ¥ ìš”ì•½
    output_buffer.append("=" * 60)
    output_buffer.append(f"ğŸ“Š ìµœì¢… ì„±ëŠ¥ ìš”ì•½ (ëª¨ë¸: {model_name})")
    output_buffer.append("-" * 60)
    output_buffer.append(f"ì´ ì´ë¯¸ì§€ ìˆ˜: {total_images}ê°œ")
    output_buffer.append(f"âœ… 'car' ìˆìŒ (O): {present_images}ê°œ")
    output_buffer.append(f"âŒ 'car' ì—†ìŒ (X): {absent_images}ê°œ")
    output_buffer.append(f"â¡ï¸ 'car'ê°€ íƒì§€ëœ ì´ë¯¸ì§€ ë¹„ìœ¨: **{presence_percentage:.2f}%**")
    output_buffer.append("-" * 60)

    # --- 5. ì½˜ì†” ì¶œë ¥ ë° ì„±ëŠ¥ ìš”ì•½ TXT íŒŒì¼ ì €ì¥ ---
    final_output = "\n".join(output_buffer)

    # ì½˜ì†”ì— ì¶œë ¥
    print(final_output)

    # TXT íŒŒì¼ë¡œ ì €ì¥ (ì„±ëŠ¥ ìš”ì•½ í¬í•¨)
    try:
        with open(summary_txt_filename, 'w', encoding='utf-8') as f:
            f.write(final_output)
        print(f"\nâœ… ë¶„ì„ ê²°ê³¼ ìš”ì•½ì´ ì„±ê³µì ìœ¼ë¡œ '{summary_txt_filename}' íŒŒì¼ì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
        print(f"TXT íŒŒì¼ ê²½ë¡œ: {os.path.abspath(summary_txt_filename)}")
    except Exception as e:
        print(f"âŒ TXT íŒŒì¼ ì €ì¥ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")


# í•¨ìˆ˜ ì‹¤í–‰
if __name__ == "__main__":
    check_car_presence_and_save(USER_CONFIG())
```
